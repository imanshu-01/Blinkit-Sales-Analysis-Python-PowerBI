# **DATA ANALYSIS PYTHON PROJECT - BLINKIT SALES ANALYSIS**


### **Import Libraries**

```python
import numpy as np              # for numerical and mathematical operations
import pandas as pd             # for data loading and manipulation
import matplotlib.pyplot as plt # for basic data visualization
import seaborn as sns           # for advanced statistical plots
```

### **Set visualization style**

```python
plt.style.use('seaborn-v0_8-darkgrid')   # graphs ka background zyada clear (looks)
sns.set_palette("husl")                  # graph ke colors & category ko alag‚Äìalag bright color
```

### **Import csv files**

```python
df=pd.read_csv("BlinkIT Grocery Data.csv")
```

### **Data Size & Sample data**

```python
print("\nüßÆ Size of Data:")
print(df.shape)                  # show number of rows and columns

print("\nüîç First 5 rows:")          
display(df.head(5))                   # display first 5 rows of the dataset
```

### **Data Information**

```python
print("\nüìã Column Information:")
print(df.info())    # Shows column names, data types, and missing values
```

### **Basic Statistics**

```python
print("\nüßÆ Basic Statistics:")
display(df.describe())           # show stats like mean, min, max, std for numeric columns
```

### **Check for missing values**

```python
print("üîé Missing Values:")

missing = df.isnull().sum()       # count missing values in each column
print(missing[missing > 0])       # show only columns that have missing data
```

```python
df.head(15)
```

### **Handle missing values**

```python
## Using Median (robust to extreme values)
df['Item Weight'] = df['Item Weight'].fillna(df['Item Weight'].median())
```

```python
df.head(15)
```

### **Several ways to fill missing values**

```python
## 1Ô∏è‚É£ Using Mean (no extreme outliers)
df['Item Weight'] = df['Item Weight'].fillna(df['Item Weight'].mean())

## 2Ô∏è‚É£ sing Mode (most frequent value)
df['Item Weight'] = df['Item Weight'].fillna(df['Item Weight'].mode()[0])

## 3Ô∏è‚É£ Forward fill (replace missing with previous row value)
df['Item Weight'] = df['Item Weight'].ffill()

## 4Ô∏è‚É£ Backward fill (replace missing with next row value)
df['Item Weight'] = df['Item Weight'].bfill()

## 5Ô∏è‚É£ Using a Custom Value
df['Item Weight'] = df['Item Weight'].fillna(10)  # replace missing values with 10
```

### **Standardize categorical columns**

```python
df['Item Fat Content'] = df['Item Fat Content'].str.lower()   # make text consistent
df['Item Fat Content'] = df['Item Fat Content'].replace({
    'lf': 'low fat', 
    'reg': 'regular'
})                                                           # standardize values

print("\n‚úÖ Data cleaning completed!")                        # confirm completion
```

#### **Sales Distribution**

```python
plt.figure(figsize=(10, 5))                 # set figure size

plt.subplot(1, 2, 1)                        # first subplot
sns.histplot(df['Sales'], bins=30, kde=True)# show sales distribution
plt.title('Sales Distribution')

plt.subplot(1, 2, 2)                        # second subplot
sns.boxplot(y=df['Sales'])                  # detect outliers
plt.title('Sales Box Plot')

plt.tight_layout()                          # adjust spacing
plt.show()                                  # display plots

## This creates a dual visualization:
## - Left: Histogram shows how sales values are distributed (frequency)
## - Right: Box plot identifies outliers and shows data spread (min, Q1, median, Q3, max)
```

## **Chart Requirements**

#### **Average Sales by Outlet Type**

```python
outlet_sales = df.groupby('Outlet Type')['Sales'] \
                 .mean() \
                 .sort_values(ascending=False)   # avg sales per outlet

plt.figure(figsize=(8, 5))                        # figure size
outlet_sales.plot(kind='bar', color='lightgreen') # bar plot
plt.title('Average Sales by Outlet Type')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)                           # rotate labels
plt.show()                                        # display plot
```

#### **Top 10 Item Types by Average Sales**

```python
top_items = df.groupby('Item Type')['Sales'].mean().sort_values(ascending=False).head(10)   # top 10 by avg sales

plt.figure(figsize=(10, 5))                        # figure size
top_items.plot(kind='bar', color='skyblue')        # bar chart
plt.title('Top 10 Item Types by Average Sales')
plt.xlabel('Item Type')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.show()                                        # display plot
```

#### **Correlation Analysis**

```python
numeric_cols = ['Item Visibility', 'Item Weight', 'Sales', 'Rating']  # numeric data
corr_matrix = df[numeric_cols].corr()                                  # correlation

plt.figure(figsize=(8, 6))                                             # figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')  # heatmap
plt.title('Correlation Heatmap')
plt.show()                                                             # display
```

#### **Key Insights & Findings**

```python
print("üéØ KEY INSIGHTS")
print("=" * 50)

# 1Ô∏è‚É£ Which outlet sells the most on average
best_outlet = df.groupby('Outlet Type')['Sales'].mean().idxmax()
best_sales = df.groupby('Outlet Type')['Sales'].mean().max()
print(f"1. Best selling outlet type: {best_outlet} (Average Sales: ‚Çπ{best_sales:.2f})")

# 2Ô∏è‚É£ Which fat content sells more
fat_preference = df.groupby('Item Fat Content')['Sales'].mean().idxmax()
print(f"2. Items with higher sales: '{fat_preference}'")

# 3Ô∏è‚É£ Does visibility affect sales?
corr_visibility_sales = df['Item Visibility'].corr(df['Sales'])
print(f"3. Item visibility and sales correlation: {corr_visibility_sales:.3f} (closer to 1 = strong effect)")

# 4Ô∏è‚É£ How consistent are ratings?
rating_std = df['Rating'].std()
print(f"4. Rating consistency (lower std = more consistent): {rating_std:.2f}")

# 5Ô∏è‚É£ Which item type sells the best
top_category = df.groupby('Item Type')['Sales'].mean().idxmax()
print(f"5. Top selling item category: {top_category}")
```

#### **Fat Content Analysis**

```python
fat_sales = df.groupby('Item Fat Content')['Sales'].agg(['mean', 'count']).round(2)   # avg & count
print("üìä Fat Content Analysis:")
display(fat_sales)

plt.figure(figsize=(8, 5))                        # size
sns.barplot(x=fat_sales.index, y=fat_sales['mean']) # bar plot
plt.title('Average Sales by Fat Content')
plt.ylabel('Average Sales')
plt.show()                                        # show
```

#### **Total Sales By Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()  # total sales
plt.pie(sales_by_fat,
        labels=sales_by_fat.index,
        autopct='%.1f%%',
        startangle=90)                                       # pie chart

plt.title('Sales By Fat Content')
plt.axis('equal')                                           # proper circle
plt.show()                                                  # display
```

#### **Total Sales By Item Type**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)   # total sales

bars = plt.bar(sales_by_type.index, sales_by_type.values)  # bar chart
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales By Item Type')
plt.xticks(rotation=45, ha='right')                        # rotate labels
plt.tight_layout()                                         # adjust layout
plt.show()                                                 # display
```

#### **Fat Content By Outlet For Total Sales**

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()   # pivot table: sales by location & fat content

# make sure both fat types exist, fill missing with 0
grouped = grouped.reindex(columns=['Regular', 'Low Fat'], fill_value=0)

ax = grouped.plot(kind='bar')                   # grouped bar chart

plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Fat Content')
plt.legend(title='Item Fat Content')
plt.xticks(rotation=45, ha='right')            # rotate labels
plt.tight_layout()                              # adjust layout
plt.show()                                     # display
```

#### **Sales Summary Statistics**

```python
#total sales
total_sales = df['Sales'].sum()
#average sales
avg_sales = df['Sales'].mean()
#No. of items 
no_of_items=df['Sales'].count()
#Average Rating
avg_rating=df['Rating'].mean()
print("\nüìä Sales Summary Statistics:")
print(f"1. Total Sales      : ${total_sales:,.0f}")
print(f"2. Avg Sales        : ${avg_sales:,.0f}")
print(f"3. No of items     : {no_of_items:,.0f}")
print(f"4. Average Rating   : {avg_rating:,.1f}")
```
